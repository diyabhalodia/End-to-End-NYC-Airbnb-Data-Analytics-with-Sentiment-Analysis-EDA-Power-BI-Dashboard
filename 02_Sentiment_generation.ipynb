```python
import numpy as np
from transformers import pipeline
import pandas as pd
```

```python
df = pd.read_excel('/content/cleaned_data.xlsx')
```

```python
sentiment_model = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")


df['sentiment'] = df['name'].astype(str).apply(lambda x: sentiment_model(x)[0]['label'])
df['sentiment_score'] = df['name'].astype(str).apply(lambda x: sentiment_model(x)[0]['score'])

df.head()
```
